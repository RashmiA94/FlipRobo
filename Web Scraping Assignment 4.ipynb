Q.1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

time.sleep(2)
```

```python
Rank=[]
Naming = []
Artist=[]
Upload_date=[]
views_billions=[]

#Extracting rank 
rank = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[1]')      
for r in rank:
        Rank.append(r.text)
time.sleep(1)
        
all_text = []
Ranking = []
for i in Rank:
    all_text.append(i.split("."))
time.sleep(2)

for i in all_text:
    Ranking.append(i[0]) 
time.sleep(1)

#Extracting name
name=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[2]')
for n in name:
        Naming.append(n.text)
time.sleep(1)

all_text3 = []
Name3 = []
for i in Naming:
    all_text3.append(i.split("["))
time.sleep(2)

for i in all_text3:
    Name3.append(i[0])
time.sleep(1)
    
#Extracting artist
artist = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[3]')      
for a in artist:
        Artist.append(a.text)
time.sleep(1)

#Extracting Upload_date
date = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[5]')      
for d in date:
        Upload_date.append(d.text)
time.sleep(1)
        
#Extracting views_billions
views = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[4]')      
for v in views:
        views_billions.append(v.text)
time.sleep(1)

# Make dataframe
Youtube_df = pd.DataFrame({'Rank':Ranking,'Name':Name3,'Artist':Artist,'Upload date':Upload_date,
                           'Views_billions':views_billions})
Youtube_df
```

Q.5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url ="https://github.com/"
driver.get(url)
time.sleep(2)
```

```python
explore=driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a")
driver.get(explore.get_attribute('href'))
```

```python
Repository_title=[]
repo_desc=[]
lang=[]

#Extracting Repository title

try:
    repo=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
    for i in repo:
        Repository_title.append(i.text)
except NoSuchElementException:
    Repository_title.append('-')
    
#Extracting Repository description

try:
    descriptions=driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/p")
    for i in descriptions:
        repo_desc.append(i.text)
except NoSuchElementException:
    repo_desc.append('-')
    
#Extracting Language used

for i in range(1,26):
    try:
        l= driver.find_element_by_xpath("//div[@class='Box']/div[2]/article[{}]/div[2]/span[1]/span[2]".format(i))
        lang.append(l.text)
    except NoSuchElementException:
        lang.append('-')
```

```python
repo_urls=[]
repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in repos:
    repo_urls.append(i.get_attribute('href'))
    
Contributors_count=[]
for i in repo_urls:
    driver.get(i)
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors_count.append(count.text)
    except NoSuchElementException:
        Contributors_count.append('-')
    
```

```python
# Make dataframe
Git_df = pd.DataFrame({'Repository_title':Repository_title,'repo_desc':repo_desc,'Contributors_count':Contributors_count,
                       'Language_used':lang})
Git_df
```

Q.8. Scrape the details of Highest selling novels.
Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/
You have to find the following details:
A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url ="https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/"
driver.get(url)
time.sleep(2)
```

```python
Book_name=[]
Author=[]
Volumes_sold=[]
Publisher=[]
Genre=[]


for i in range(0,100):
    
    # Extracting book name
    
    book = driver.find_elements_by_id("table-cell-10943-{}-1".format(i))
    for b in book:
        Book_name.append(b.text)
     
    
    # Extracting author name
    
    author = driver.find_elements_by_id("table-cell-10943-{}-2".format(i))
    for a in author:
        Author.append(a.text)
      
    
    # Extracting volumes sold
    
    volume = driver.find_elements_by_id("table-cell-10943-{}-3".format(i))
    for v in volume:
        Volumes_sold.append(v.text)
        
    # Extracting publisher
    
    publisher = driver.find_elements_by_id("table-cell-10943-{}-4".format(i))
    for p in publisher:
        Publisher.append(p.text)
        
        
    # Extracting genre
    
    genre = driver.find_elements_by_id("table-cell-10943-{}-5".format(i))
    for g in genre:
        Genre.append(g.text)
        
# Make dataframe
guardian_df = pd.DataFrame({'Book name':Book_name,'Author name':Author,'Volumes sold':Volumes_sold,
                       'Publisher':Publisher,'Genre':Genre})
guardian_df
```

Q.9. Scrape the details most watched tv series of all time from imdb.com.

You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url ="https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(2)
```

```python
Name=[]
Year_span=[]
Genre1=[]
Run_time=[]
Ratings=[]
Votes=[]

# Extracting Name

name1 = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a')
for n in name1:
    Name.append(n.text)

# Extracting Year span
year = driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')
for y in year:
    Year_span.append(y.text)

# Extracting Genre
genre1 = driver.find_elements_by_xpath('//span[@class="genre"]')
for g in genre1:
    Genre1.append(g.text)

# Extracting Run time   
run = driver.find_elements_by_xpath('//span[@class="runtime"]')
for r in run:
    Run_time.append(r.text)
    
# Extracting Ratings    
rating = driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]/span[2]')
for s in rating:
    Ratings.append(s.text)

# Extracting Votes
vote = driver.find_elements_by_xpath('//span[@name="nv"]')
for v in vote:
    Votes.append(v.text)
    
# Make dataframe
imdb_df = pd.DataFrame({'Name':Name,'Year span':Year_span,'Genre':Genre,
                       'Run time':Run_time,'Ratings':Ratings,'Votes':Votes})
imdb_df
```
