# Q1.Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location.

```python
driver=webdriver.Chrome("chromedriver.exe")
url="https://www.naukri.com/"
driver.get(url)
search_job=driver.find_element_by_xpath('//input[@id="qsb-keyword-sugg"]')
search_job.send_keys("Data Analyst")
search_location=driver.find_element_by_xpath('//input[@id="qsb-location-sugg"]')
search_location.send_keys("Bangalore")
search_btn=driver.find_element_by_xpath('//div[@class="search-btn"]').click()
```

```python
title_tag=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
title=[]
for i in title_tag:
    title.append(i.text)
title[:10]
```

```python
company_tag=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
company=[]
for i in company_tag:
    company.append(i.text)
company[:10]
```

```python
location_tag=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
location=[]
for i in location_tag:
    location.append(i.text)
location[:10]
```

```python
experience_tag=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
experience=[]
for i in experience_tag:
    experience.append(i.text)
experience[:10]
```

```python
df=pd.DataFrame({"job title":title[:10],
                "job location":location[:10],
                "job company":company[:10],
                "job experience":experience[:10]})
df
```

# Q2.Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location.

```python
driver=webdriver.Chrome("chromedriver.exe")
url="https://www.naukri.com/"
driver.get(url)
```

```python
search_job=driver.find_element_by_xpath('//input[@id="qsb-keyword-sugg"]')
search_job.send_keys("Data Scientist")
search_location=driver.find_element_by_xpath('//input[@id="qsb-location-sugg"]')
search_location.send_keys("Bangalore")
search_btn=driver.find_element_by_xpath('//div[@class="search-btn"]').click()
```

```python
title_tag1=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
title1=[]
for i in title_tag1:
    title1.append(i.text)
title1[:10]
```

```python
location_tag1=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
location1=[]
for i in location_tag1:
    location1.append(i.text)
location1[:10]
```

```python
company_tag1=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
company1=[]
for i in company_tag1:
    company1.append(i.text)
company1[:10]
```

```python
urls1=[]
for i in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]'):
    urls1.append(i.get_attribute('href'))
urls1[:10]
```

```python
import time
from selenium.common.exceptions import NoSuchElementException 
job_desc1=[]
for i in urls1[:10]:
    driver.get(i)
    time.sleep(3)
    try:
        
        job_desc=driver.find_element_by_xpath("//section[@class='job-desc']")
        job_desc1.append(job_desc.text)
    except NoSuchElementException:
        job_desc1.append('-')
```

```python
job_desc1[:10]
```

```python
print(len(job_desc1))
```

```python
df1=pd.DataFrame({"job title":title1[:10],
                "job location":location1[:10],
                "job company":company1[:10],
                "job description":job_desc1[:10]})
df1
```
